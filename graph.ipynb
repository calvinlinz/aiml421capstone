```python
import matplotlib.pyplot as plt
import numpy as np

# Data
epochs = np.arange(1, 11)
losses = [1.4775, 1.0897, 1.0053, 0.8376, 0.6145, 0.3642, 0.1946, 0.0777, 0.0370, 0.0164]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(epochs, losses, 'b-o', linewidth=2, markersize=8)

# Customize the plot
plt.title('Training Loss Over Epochs - CNN Baseline', fontsize=14, pad=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Set y-axis limits
plt.ylim(0, max(losses) + 0.2)

# Mark initial and final points
plt.annotate(f'Initial Loss: {losses[0]:.4f}', 
           xy=(1, losses[0]), 
           xytext=(2, losses[0]+0.1),
           arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate(f'Final Loss: {losses[-1]:.4f}', 
           xy=(len(losses), losses[-1]), 
           xytext=(7, 0.3),
           arrowprops=dict(facecolor='black', shrink=0.05))

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
